#### **solving** 

With all the necessary information set up, we can solve the problem.
We loop over all elements in the structure, construct the element stiffness matrices, rotate these to align with the global coordinate system, and, finally, assemble the element stiffness matrices into the ```global_stiffness_matrix```.
(We may also store some information for post-processing purposes.)
Make sure this step is implemented generically (and not for this specific example), since we will want to reuse the code for other meshes.

````python
# BEGIN SOLUTION NO PROMPT
# ----------------------------------------------------------------------------------------------------------#
#                                                ASSEMBLE K
# ----------------------------------------------------------------------------------------------------------#
# intialize 
global_stiffness_matrix = np.zeros((no_global_dofs, no_global_dofs))

elm_length_vector = np.zeros(no_elements)
struct_element_rotation_matrix = []

for e in range(no_elements):
    node1e = nodal_connectivities[e, 0]
    node2e = nodal_connectivities[e, 1]
    
    elm_orientation_vector = nodal_positions[node2e] - nodal_positions[node1e]
    elm_orientation_angle = np.arctan2(elm_orientation_vector[1], elm_orientation_vector[0])
    elm_length = np.linalg.norm(elm_orientation_vector)
    elm_length_vector[e] = elm_length

    axial_stiffness_matrix = A * Ey / elm_length * np.array([[1, -1],
                                                            [-1, 1]])
    
    rotation_matrix = np.array([[np.cos(elm_orientation_angle), np.sin(elm_orientation_angle), 0, 0],
                              [0, 0, np.cos(elm_orientation_angle), np.sin(elm_orientation_angle)]])
    
    # save rotation matrix for later on
    element_rotation = {'a': rotation_matrix}
    struct_element_rotation_matrix.append(element_rotation)

    element_stiffness_matrix = rotation_matrix.T @ axial_stiffness_matrix @ rotation_matrix

    global_stiffness_matrix[np.ix_(element_dof_connectivity[e], element_dof_connectivity[e])]+= element_stiffness_matrix
# END SOLUTION
""" # BEGIN PROMPT
# ----------------------------------------------------------------------------------------------------------#
#                                                ASSEMBLE K
# ----------------------------------------------------------------------------------------------------------#2
# intialize the global_stiffness_matrix to the correct size filled with zeros:
global_stiffness_matrix = ...

# compute the elements of the global_stiffness_matrix generically
# making use of the useful variables that we defined in the beginning
# and add their contributions to the global_stiffness_matrix.
# write your code here     
# ...

""" # END PROMPT
````
