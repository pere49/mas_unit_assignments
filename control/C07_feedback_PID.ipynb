<b> C 9.1: Root locus </b>

Given the following third-order plant
\begin{equation*}
G(s) = \frac{1}{(s+10)(s+2)^2},
\end{equation*}
please study the influence of the gain $k$ of a PID controller $K(s)=k(1+\frac{1}{T_I s}+T_D s)$ with $T_I =\frac{6}{5}$ and $T_D =\frac{1}{6}$. To do so, plot the root locus plot of the open loop system.

> **Activity:**  
> Can you decrease the settling time by increasing $k$?  
> Can you improve the settling time and the overshoot at the same time by changing $k$?  
> Is the system stable for a large control gain ($k \rightarrow \infty$)?

```python
# your code goes here
s = ct.tf('s')
T_i = 1.2
T_d = 1/6
k = (1 + 1/T_i*s + T_d*s)
G = 1/(s*(s+1)*(s+2))
L = k*G
ct.root_locus(L)
plt.title('Root Locus of the Given System')
plt.show()
```
