```python
# TODO: toggle
#system matrices
A = np.array([[0,1],[0,-10]]);
B = np.array([[0],[1]]);
C = np.array([[1,0]]);
D = np.array([[0]]);

#step response requirements
#maxmimum overshot
M_p = 0.25;

#maximum settling time
ts = 0.115;

#compute damping ratio and natural frequency using the formulas from the lecture
z = sp.symbols('z');
expr = sp.exp(-sp.pi*z/sp.sqrt(1-z*z))-M_p;
sol_zeta = sp.solve(expr);
zeta = sol_zeta[0];

omega_n = 4.6/(zeta*ts);

#compute desired closed-loop poles
p = np.roots([1, 2*zeta*omega_n, omega_n*omega_n]);

#compute state feedback controller
K = ct.place(A,B,p);

#compute step response of closed-loop system 
sys_cl = ct.ss(A-B*K,B,C,D);

t_step,y_step = ct.step_response(sys_cl, T=0.16, return_x=False)
plt.plot(t_step,y_step,label='$y$')
plt.xlabel("seconds")
plt.title("step response")
plt.legend()
plt.grid()
plt.show()
```
