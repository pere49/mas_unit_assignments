```python
# TODO: toggle
#system matrices for observer 1
A_cl1 = np.block([
   [A-np.matmul(B,K),np.matmul(B,K)],
   [np.zeros((4,4)),A-np.matmul(L1,C)]
                ]);
B_cl1 = np.block([
    [B],
    [np.zeros((4,1))]
                ]);
C_cl1 = np.block([
    [C,np.zeros((2,4))],
    [np.zeros((2,4)),C]
                ]);
D_cl1 = np.block([
    [D],
    [D]
        ]);

#system matrices for observer 2
A_cl2 = np.block([
    [A-np.matmul(B,K),np.matmul(B,K)],
    [np.zeros((4,4)),A-np.matmul(L2,C)]
                      ]);
B_cl2 = np.block([
    [B],
    [np.zeros((4,1))]
                ]);
C_cl2 = np.block([
    [C,np.zeros((2,4))],
    [np.zeros((2,4)),C]
                ]);
D_cl2 = np.block([
    [D],
    [D]
        ]);

#create state-space objects
sys1 = ct.ss(A_cl1,B_cl1,C_cl1,D_cl1);
sys2 = ct.ss(A_cl2,B_cl2,C_cl2,D_cl2);

#initial condition
x_0 = np.array([2, 0.5, 1, 0, 1, 0.1, 0, 0]);

#compute output and error dynamics
t1,y1 = ct.initial_response(sys1,T=None,X0=x_0);
t2,y2 = ct.initial_response(sys2,T=None,X0=x_0);

#plots
plt.plot(t1,y1[0],label='$y_{1,1}$')
plt.plot(t1,y2[0],label='$y_{2,1}$')
plt.xlabel("seconds")
plt.grid()
plt.legend()
plt.show()

plt.plot(t1,y1[1],label='$y_{1,2}$')
plt.plot(t1,y2[1],label='$y_{2,2}$')
plt.xlabel("seconds")
plt.grid()
plt.legend()
plt.show()

plt.plot(t1,y1[2],label='$e_{1,1}$')
plt.plot(t1,y2[2],label='$e_{2,1}$')
plt.xlabel("seconds")
plt.grid()
plt.legend()
plt.show()

plt.plot(t1,y1[3],label='$e_{1,2}$')
plt.plot(t1,y2[3],label='$e_{2,2}$')
plt.xlabel("seconds")
plt.grid()
plt.legend()
plt.show()
```

<i> Discussion on the performance of the two observers: </i> 
An observer should always have faster dynamics than the closed-loop system such that the error between the measured and the estimated state is minimized. The poles of observer 1 are closer to those of the closed-loop system than the poles of observer 2. In particular, we find that observer 1 is not fast enough.<br>
Hence, the second observer gives better performance since the error dynamics $e_{2,1}(t)$ and $e_{2,2}(t)$ go much faster to zero than for the first observer.
