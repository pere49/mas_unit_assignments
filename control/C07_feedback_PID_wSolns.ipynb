```python
s = ct.tf('s')
T_I = 6/5
T_D = 1/6
K = (1+ 1/(T_I * s) + T_D * s)
G = 1/ ((s+10) * (s+2)**2)
L = G * K
plt.figure()
ct.root_locus_plot(L)
plt.show()
```

> We can decrease the settling time by increasing the control gain $k$. The reason is that the poles move further away from the imaginary axis as $k$ increases.  
> We can not decrease the settling time and overshoot simultaneously. This is due to the observation that as $k$ increases, hte angle of the pole locations with respect to the negative real axis increases.  
> The system is stable. This is because as the control gain $k$ increases to infinity, all the poles lie in the left half-space.

> The zeros of $K(s)*G(s)$ correspond to the zeros of $s^2 + s/T_D + 1/(T_D*T_I)$. Hence, $z_1 z_2 = 1/(T_D*T_I)$, and $z_1 + z_2 = -1/T_D$. Therefore, $T_I = -(z_1+z_2)/(z_1 z_2)$, and $T_D = -1/(z_1+z_2)$. We can then plot the root locus for different scenarios.

```python
# specify the list of zeros
z_1_list = [-1, 1, -5, -2]
z_2_list = [-5, -5, -20, -2]

# set the transfer function
s = ct.tf('s')
G = 1/((s + 10) * (s + 2)**2)

for i, (z1, z2) in enumerate(zip(z_1_list, z_2_list), start=1):
    T_D = -1.0 / (z1 + z2)
    T_I = -(z1 + z2) / (z1 * z2)
    K = 1 + 1/(T_I * s) + T_D * s
    L = G * K

    plt.figure()
    ct.root_locus_plot(L)
    ax = plt.gca()
    ax.set_title(f'Root locus (zeros={z1}, {z2}; T_D={T_D:.3g}, T_I={T_I:.3g})')
    ax.margins(0.02)
    plt.show()
```

> We find that the root locus corresponding to the case where the zeros are $(-2,-2)$ is desirable. This is because the poles are away from the imaginary axis, which implies stability when we choose a large gain. we will further select the gain to achieve a balance between a small settling time and a small overshoot.

```python
z1 = -2
z2 = -2
T_D = -1.0 / (z1 + z2)
T_I = -(z1 + z2) / (z1 * z2)
K = 1 + 1/(T_I * s) + T_D * s
L_raw = G * K

K_u = 100
L = K_u * L_raw

closed_loop = ct.feedback(L,1)
T, y = ct.step_response(closed_loop)
plt.figure()
plt.plot(T, y)
plt.xlabel('Time (s)')
plt.ylabel('Output $y$')
plt.grid()
plt.show()
```
