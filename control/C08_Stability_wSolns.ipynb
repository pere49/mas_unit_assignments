<b> C 10.2: Bode plot </b>

Consider the following continuous-time linear time-invariant system 
\begin{align*}
\dot{x}(t) &= \begin{bmatrix} -2& -6 \\ 0& -1\end{bmatrix} x(t)+\begin{bmatrix} 1\\0\end{bmatrix}u_{ac}(t),\\
y(t) &= \begin{bmatrix}1&0 \end{bmatrix}x(t)+u_{ac}(t)
\end{align*}
where $u_{ac}$ is the input to the system and also the output of an actuator with transfer function $G_{ac}=\frac{U_{ac}(s)}{U(s)}=\frac{100}{(s+10)^2}$.

<blockquote>
<b>Activity:</b> Generate the Bode plot of the open-loop system consisting of the plant and the actuator.
    
<i>Hint: Derive the transfer function G of the plant and represent the entire system as G*G_ac<i>
</blockquote>

```python
# your code goes here
```

<blockquote>
<b>Solution:</b> You can check your answer with the solution by clicking on the "..." below.
</blockquote>

```python
# system matrices
A = np.array([[-2, -6],[0, -1]])
B = np.array([[2],[0]])
C = np.array([[1,0]])
D = np.array([1])

# build a continuous-time state-space object
sys_ct = ct.ss(A,B,C,D)

# transfer function of the actuator
G_ac = ct.tf([100],[1,20,100])
L = sys_ct * G_ac

plt.figure()
ct.bode_plot(L)
```

<blockquote>
<b>Activity:</b> Read off the gain and phase margin, and their corresponding frequencies . Calculate the gain and phase margin using the Python control package to validate your answer.
</blockquote>

```python
# your code goes here
```

<blockquote>
<b>Solution:</b> You can check your answer with the solution by clicking on the "..." below.
</blockquote>

```python
GM, PM, GM_frequency, PM_frequency = ct.margin(L)
print(GM, GM_frequency)
print(PM, PM_frequency)

freq = np.logspace(-2, 3, 1000)
ct.bode_plot(L, freq, display_margins=True)
```

```python
# for a DC-gain equal one we need a gain of 20*30=600, multiply (s+20) and (s+30) to get (s^2+50s+600)
measurement = ct.tf([600],[1,50,600])
L_op = measurement * sys_ct * G_ac
plt.figure()
ct.bode_plot(L_op)
plt.show()
```
