```python
A = np.array([[-1, 2, -3, 0],[0, -2, 6, 4],[0, 0, -4, 4],[0, 0, 0, -6]])
B = np.array([[0],[1],[0],[0]])
W_c = ct.ctrb(A,B)
np.linalg.matrix_rank(W_c)
# given that the rank is only 2 the system is not controllable
```

Next, we want to bring the system to the controllability staircase form to find out which states are controllable and which ones are not.

```python
print(W_c)
T1 = W_c[:,0:np.linalg.matrix_rank(W_c)]
print(T1)
T2 = np.array([[0, 0],[0, 0],[1, 0],[0, 1]])
print(T2)
T = np.hstack((T1,T2))
print(T)
```

```python
A_scf = np.dot(np.dot(np.linalg.inv(T),A),T)
B_scf = np.dot(np.linalg.inv(T), B)
print(A_scf)
print(B_scf)
A_unc = A_scf[2:4,2:4]
np.linalg.eig(A_unc)
# The eigenvalues are negative. Therefore the uncontrollable subspace is stable and all uncontrollable states decay to zero. 
# That means the system is stabilizable. Note, that we could have also calculated the eigenvalues of A 
# which are all negative which means the system has to be stabilizable.
```
