```python
# TODO: toggle
#system matrices
A = np.array([[-2, -6],[0, -1]])
B = np.array([[1],[0]])
C = np.array([[1,0]])
D = np.array([1])

#sampling time [sec]
T_s = 0.1

#build a continuous-time state-space object and discretize it
sys_ct = ct.ss(A,B,C,D)
sys_dt = ct.StateSpace.sample(sys_ct,T_s,method='zoh')
sys_dt
```
